```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('paper')
%matplotlib inline
```

```python
df = pd.read_csv('weiss18table.txt', index_col=False, delimiter=' ', skipinitialspace=True, skiprows=29, names=['KOI-S', 'KIC', 'KOI-P', 'Mstar', 'Rstar', 'CDPP', 'b', 'Per', 'Rp', 'e_Rp', 'T', 'RpAvg', 'Prat', 'Delta'])
df.head()
```

```python
Np = {s:(df['KOI-S']==s).sum() for s in df['KOI-S'].unique()}
df['Nplanets'] = [Np[s] for s in df['KOI-S']]
df.tail(20)
```

```python
df['mu'] = 3*(2/df['Delta']*(df['Prat']-1)/(df['Prat']+1))**3
df['Delta1/3'] = (df['Prat']-1)/(df['mu']/3)**(1/3)
df['S'] = (1-df['Prat']**(-2/3))/df['mu']**(1/4)
df['ecritquad'] = 1/df['Prat']/9.24*df['S']**2
df
```

```python
fig, ax = plt.subplots(figsize=(8,6))
df3 = df[df['Nplanets'] > 2]
df2 = df[df['Nplanets'] == 2]
ax.hist(df2['S'], bins=np.linspace(0, 12, 25), fill=False, edgecolor='b', linewidth=4, label='2 planet systems')
ax.hist(df3['S'], bins=np.linspace(0, 12, 25), fill=False, edgecolor='r', linewidth=4, label='3+ planet systems')
ax.set_xlabel('Dynamical Spacing $S$')
ax.set_ylabel('Number of Adjacent Planet Pairs')
ax.legend()
fig.savefig('dynspacing.pdf')
```
